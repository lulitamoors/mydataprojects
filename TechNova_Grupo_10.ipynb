```python
import matplotlib.pyplot as plt
```

```python
#Filtramos los empleados que pertenecen al departamento de Recursos Humanos.
TrainingDB_RH = TrainingDB[TrainingDB['Department'] == 'Recursos Humanos']

# Contamos y mostramos la cantidad de empleados de Recursos Humanos por programa
# de capacitación.
Cantidad_empleados_RH = TrainingDB_RH['Training_Program'].value_counts()
print(Cantidad_empleados_RH)
```

```python
#Crear y mostrar el resultado en gráfico de barras.
Cantidad_empleados_RH.plot(kind='bar', color='pink')
plt.title('Cantidad de empleados de Recursos Humanos por programa de capacitación')
plt.xlabel('Programas de capacitación')
plt.ylabel('Cantidad de empleados')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Sacamos y mostramos el promedio de rendimiento de antes y después de la
# capacitación
Promedio_antes = TrainingDB['Performance_Before_Training'].mean()
Promedio_despues = TrainingDB['Performance_After_Training'].mean()
print(f'Promedio antes de la capacitación: {Promedio_antes}')
print(f'Promedio después de la capacitación: {Promedio_despues}')
```

```python
#Graficamos el resultado
plt.bar(['Antes de Capacitación', 'Después de capacitación'], [Promedio_antes, Promedio_despues], color=['mediumorchid', 'lightskyblue'])
plt.title('Comparación del rendimiento antes y después de la capacitación')
plt.ylabel('Promedio de rendimiento')
plt.show()
```

```python
#Contamos y mostramos el número de empleados que hay por departamento
Conteo_departamento = TrainingDB['Department'].value_counts()
print(Conteo_departamento)
```

```python
#Graficamos y mostramos el resultado
Conteo_departamento.plot(kind='bar', color='aqua')
plt.title('Número de empleados por departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de empleados')
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Contamos y mostramos cuántos empleados recibieron promoción después de la
# capacitación
Conteo_promocion = TrainingDB['Promotion_After_Training'].value_counts()
print(Conteo_promocion)
```

```python
#Graficamos el pie y mostramos el resultado
Conteo_promocion.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['hotpink', 'greenyellow'])
plt.title('Empleados con promoción después de la capacitación')
plt.ylabel('')
plt.show()
```

```python
# Graficamos la relación entre satisfacción y rendimiento después de la
# capacitación
plt.scatter(TrainingDB['Satisfaction_Score'], TrainingDB['Performance_After_Training'], color='purple')
plt.title('Relación entre satisfacción y rendimiento después de la capacitación')
plt.xlabel('Puntaje de satisfacción')
plt.ylabel('Rendimiento después de la capacitación')
plt.grid(True)
plt.show()
```

```python
# Agrupamos por programa de capacitación y calculamos el promedio de
# rendimiento después de la capacitación
Rendimiento_capacitacion = TrainingDB.groupby('Training_Program')['Performance_After_Training'].mean()

#Ordenamos los programas por rendimiento y seleccionamos los 3 mejores
Top_capacitaciones = Rendimiento_capacitacion.sort_values(ascending=False).head(3)

#Mostramos el resultado
print(Top_capacitaciones)
```

```python
#Creamos y mostramos la gráfica con el resultado
plt.plot(Top_capacitaciones.index, Top_capacitaciones.values, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.title('Top 3 programas con mejor rendimiento después de la capacitación')
plt.xlabel('Programas de capacitación')
plt.ylabel('Rendimiento promedio')
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.show()
```

```python
#Creamos listas con los rendimientos antes y después de la capacitación
Rendimiento_antes = TrainingDB['Performance_Before_Training'].tolist()
Rendimiento_despues = TrainingDB['Performance_After_Training'].tolist()

#Creamos una lista para almacenar las diferencias
Rendimiento_diferencia = []

#Usamos un bucle for para calcular la diferencia entre antes y después
for Antes, Despues in zip(Rendimiento_antes, Rendimiento_despues):
    Diferencia = Antes - Despues
    Rendimiento_diferencia.append(Diferencia)

#Mostramos las diferencias
print("Diferencias de rendimiento:", Rendimiento_diferencia)
```

```python
#Graficamos la comparación
plt.plot(Rendimiento_antes, label='Antes de Capacitación', color='red', marker='o')
plt.plot(Rendimiento_despues, label='Después de Capacitación', color='green', marker='o')
plt.title('Comparación de rendimiento antes y después de la capacitación')
plt.xlabel('Empleados')
plt.ylabel('Rendimiento')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Filtramos empleados con rendimiento alto (>= 88) después de la capacitación
Alto_rendimiento = list(filter(lambda x: x >= 88, TrainingDB['Performance_After_Training']))

#Filtramos empleados con rendimiento bajo (<= 78) después de la capacitación
Bajo_rendimiento = list(filter(lambda x: x <= 78, TrainingDB['Performance_After_Training']))

#Mostramos los resultados
print(f"Empleados con rendimiento alto (>=88): {len(Alto_rendimiento)}")
print(f"Empleados con rendimiento bajo (<=78): {len(Bajo_rendimiento)}")
```

```python
#Graficamos los resultados
plt.bar(['Rendimiento Alto', 'Rendimiento Bajo'], [len(Alto_rendimiento), len(Bajo_rendimiento)], color=['cornflowerblue', 'violet'])
plt.title('Empleados con rendimiento alto y bajo después de la capacitación')
plt.ylabel('Cantidad de empleados')
plt.show()
```

```python
# Creamos subconjuntos con empleados que recibieron promoción y los que no
Promovido = TrainingDB[TrainingDB['Promotion_After_Training'] == 'Yes']
No_promovido = TrainingDB[TrainingDB['Promotion_After_Training'] == 'No']

# Unimos ambos subconjuntos
Datos_combinados = pd.concat([Promovido, No_promovido])

# Graficamos el número de empleados promovidos y no promovidos
Estado_promocion = Datos_combinados['Promotion_After_Training'].value_counts()
Estado_promocion.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Distribución de promociones después de la capacitación')
plt.ylabel('')
plt.show()
```

```python
#Filtramos empleados con rendimiento superior a 88 después de la capacitación
Emp_alto_desempeño = TrainingDB[TrainingDB['Performance_After_Training'] >= 88]

#Agrupamos por departamento y contar el número de empleados
Dep_alto_desempeño = Emp_alto_desempeño.groupby('Department')['Employee_ID'].count()

#Mostramos el resultado
print(Dep_alto_desempeño)
```

```python
#Definimos una lista de colores para cada departamento
Colores = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#8A2BE2', '#FF69B4']

#Graficamos el resultado con diferentes colores
plt.bar(Dep_alto_desempeño.index, Dep_alto_desempeño.values, color=Colores[:len(Dep_alto_desempeño)])
plt.title('Empleados con rendimiento superior o igual a 88 por departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de empleados')
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y')
plt.show()
```
